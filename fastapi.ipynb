```python
#todo list
from fastapi import FastAPI
from pydantic import BaseModel

app= FastAPI(
    title ="FastAPI KT",
    description="Fast API complet etutorial ",
    version="1.0.0"
)
task_list=[]

class Task(BaseModel):
    id : int
    name: str
    completed: bool

@app.post("/tasks")
def add_task(new_task: Task):
    task_list.append(new_task)
    return{
        "message":"task added",
           "data": new_task   }
    
@app.get("/tasks")
def get_tasks():
    return task_list

@app.get("/tasks/{task_id}")
def get_single_task(task_id: int):
    for task in task_list:
        if task.id ==task_id:
            return task
    return{"error":"task not found"}

@app.put("/tasks/{task_id}")
def update_task(task_id:int, new_data:Task):
    for index, task in enumerate(task_list):
        if task.id ==task_id:
            task_list[index]=new_data

            return{
                "message":"Task updated",
                "data":new_data
            }
    return{"error": "task not found"}

@app.delete("/tasks/{task_id}")
def delete_task(task_id:int):
    for index, task in enumerate(task_list):
        if task.id ==task_id:
            task_list.pop(index)

            return{
                "message":"Task deleted"
        
            }
    return{"error": "task not found"}
```
